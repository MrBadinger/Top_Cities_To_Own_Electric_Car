```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load regs and incentives list csv
reg_incent_toload = "../Resources/ev_incentives_by_state.csv"

reg_incent_df = pd.read_csv(reg_incent_toload)
reg_incent_df
```

```python
#reg_incent_df["Registrations"] = reg_incent_df["Registrations"].astype(int)
```

```python
incentives = reg_incent_df["Incentives"].tolist()
registrations = reg_incent_df["Registrations"].tolist()
#registrations

```

```python

plt.scatter(incentives, registrations)
plt.show()

#plt.hist(registrations.index, registrations)
#plt.xticks(rotation='vertical')
#plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(18,3))

color = 'tab:red'
ax1.set_xlabel("State")
ax1.set_ylabel("Registrations", color=color)
ax1.plot(reg_incent_df["State Code"], registrations, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Incentives', color=color)
ax2.plot(reg_incent_df["State Code"], incentives, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
